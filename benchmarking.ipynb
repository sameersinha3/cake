```python
prompts = [
    "What is the capital of Japan?",
    "Write a short story about an astronaut stranded on Mars.",
    "The first three numbers in a sequence are 3, 6, and 9...",
    "Write a Python function to sort a list using quicksort.",
    "What is a black hole?",
    "You have a 3-gallon and a 5-gallon jug...",
]
```

```python
import time
llm = LLM("meta-llama/Llama-2-7b-chat-hf", kv_cache_dtype="fp16", max_model_len=8192, enforce_eager=True, device='cpu')
sampling_params = SamplingParams(temperature=0.7, max_tokens=200)


start_time = time.time()
for prompt in prompts:
    output = llm.generate(prompt, sampling_params)
end_time = time.time()

print(f"Generated in {end_time - start_time:.2f} seconds")
print(f"Average Time per response: {(end_time - start_time)/len(prompts):.2f} seconds")
```
